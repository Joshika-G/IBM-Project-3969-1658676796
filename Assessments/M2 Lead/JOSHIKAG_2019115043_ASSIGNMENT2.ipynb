##Univariate Analysis

```python
sns.countplot(x = "Exited" , data = df)
plt.xlabel("0: Customer with bank  1: Customer exited")
plt.ylabel("Count of customers")
plt.title("Bank Customers Visualization")
plt.show()
print()

sns.countplot(x = "Tenure" , data = df)
plt.xlabel("Tenure")
plt.ylabel("Count")
plt.title("Bank Customers Tenure Visualization")
plt.show()
print()

plt.scatter(df.Balance, df.CreditScore)
plt.xlabel("Scatterplot Balance")
plt.ylabel("Credit Score")
plt.title("Scatterplot for Balance and Credit score")
plt.show()
print()

plt.scatter(df.CreditScore, df.EstimatedSalary)
plt.xlabel("Credit Score")
plt.ylabel("Estimated Salary")
plt.title("Scatterplot for Credit Score and Estimated Salary")
plt.show()
print()
```

##Bi-variate Analysis

```python

sns.barplot(x = "Exited", y = "Age", data = df)
plt.xlabel("Exited")
plt.ylabel("Age")
plt.title("Visualization")
plt.show()
print()

sns.boxplot(x = "Geography", y = "EstimatedSalary", data = df, palette = 'rainbow')
plt.xlabel("Geography")
plt.ylabel("Estimated Salary")
plt.title("Boxplot Visualization")
plt.show()
print()

sns.violinplot(x = "Geography", y = "CreditScore", data = df, palette = 'rainbow')
plt.xlabel("Geography")
plt.ylabel("Credit Score")
plt.title("Boxplot Visualization")
plt.show()
print()
```

##Multi-variate Analysis 

```python
#multivariate viz
sns.pairplot(df)
plt.title("Multivariate Visualization")
plt.show()
print()
```

##Descriptive Statistics

```python
#Descriptive statistics
a = df[['RowNumber', 'CustomerId', 'CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Exited']].copy()
print("Median\n",np.median(a,axis=0))
print("Mode\n",stats.mode(a,axis=0))
print("Variance\n",np.var(a,axis=0))
print("Standard Deviation\n",np.std(a,axis=0))
print("Kurtosis\n",stats.kurtosis(a,axis=0))
print("Skewness\n",stats.skew(a,axis=0))
```
